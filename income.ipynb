```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='median')
df['relative_per_capita_income']=imp.fit_transform(df[['relative_per_capita_income']])
```

```python
imp=SimpleImputer(strategy='most_frequent')
df['notes'] = imp.fit_transform(df[['notes']]).ravel()
```

```python
cat_cols = ['state', 'state_size', 'fiscal_year', 'notes']
ohe = OneHotEncoder()
cat_encoded = ohe.fit_transform(df[cat_cols]).toarray()

# 3. Normalize numeric variables
num_cols = ['relative_per_capita_income']
scaler = MinMaxScaler()
num_scaled = scaler.fit_transform(df[num_cols])

# 4. Combine features
import numpy as np
X = np.hstack((num_scaled, cat_encoded))

# 5. Apply K-means
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X)
```

# Elbow Methos

```python
wcss = []  # Within-Cluster Sum of Squares

# Try k from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```
